```python
import pandas as pd
import re
```

```python
df=pd.read_excel('data.xlsx')
```

```python
df
```

```python
def cleanup(x):
    try:
        return eval(x)
    except:
        data = re.split(r'[,+ .]', x)
        cleaned = [x for x in data if x not in ['','(automated)','or']]
        return cleaned
```

```python
df['Corrected_Attribute']=df['Corrected_Attribute'].map(cleanup)
```

```python
count=0
for i in df['Corrected_Attribute']:
    if i[0] in ['Unkown','Unkknown']:
        print(count)
    count+=1
```
